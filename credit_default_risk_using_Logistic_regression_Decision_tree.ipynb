```python
from sklearn.metrics import roc_curve, auc

def _plot_roc_curve(fpr, tpr, thres, auc):
    plt.figure(figsize = (10, 8))
    plt.plot(fpr, tpr, 'b-', color='darkorange', lw=2, linestyle='--', label='ROC curve (area = %0.2f)'%auc)
    plt.plot([0, 1], [0, 1], '--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.title('ROC Curve')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

### Pre-pruning (Hyperparameter Tuning)

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
DT_classifier_pre_pruning = DecisionTreeClassifier()
DT_classifier_pre_pruning.fit(X_train, y_train.ravel())

y_pred = DT_classifier_pre_pruning.predict(X_train)
```

```python
DT_classifier_pre_pruning.get_params()
```

## Firstly, we used RandomSearch to find the optimal interval of hyperparameters 

```python
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
max_features.append(None)
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 20, num = 6)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
criterion = ['gini', 'entropy', 'log_loss']
# Create the random grid
random_grid = {'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'criterion': criterion}
print(random_grid)
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
dt_random = RandomizedSearchCV(estimator = DT_classifier_pre_pruning, param_distributions = random_grid, n_iter = 100, cv = 4, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
dt_random.fit(X_train, y_train)
```

```python
dt_random.best_params_
```

```python
y_pred_random = dt_random.predict(X_train)
print(confusion_matrix(y_train,y_pred_random))
print(classification_report(y_train,y_pred_random))
print('Decision Tree accuracy: ', accuracy_score(y_train, y_pred_random).round(2))
```

## After identifying the interval, I used GridSearch to find the optimal combination of hyperparameter

```python
param_grid = {'min_samples_split': [5,6,7],
              'min_samples_leaf': [3,4,5],
              'max_depth': [4,5,6]}

dt_grid = GridSearchCV(estimator = DT_classifier_pre_pruning, param_grid = param_grid, cv = 4, n_jobs = -1, verbose = 2)

dt_grid.fit(X_train, y_train)
dt_grid.best_params_
```

```python
y_pred_grid = dt_grid.predict(X_train)
print(confusion_matrix(y_train,y_pred_grid))
print(classification_report(y_train,y_pred_grid))
print('Decision Tree accuracy: ', accuracy_score(y_train, y_pred_grid).round(2))
```

## Post Pruning

```python
#DT_classifier = DecisionTreeClassifier(max_depth= 4, min_samples_leaf= 3, min_samples_split= 5, criterion='entropy')
DT_classifier = DecisionTreeClassifier()
DT_classifier.fit(X_train, y_train.ravel())

path = DT_classifier.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
ccp_alphas
```

### Because ccp_alpha must be greater than or equal to 0 so I removed the negative elements

```python
new_ccp_alphas = np.delete(ccp_alphas, 1)
```

```python
fig, ax = plt.subplots()
fig = plt.figure(figsize =(12, 7))
ax.plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
```

```python
clfs = []
for ccp_alpha in new_ccp_alphas:
    DT_classifier = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    DT_classifier.fit(X_train, y_train.ravel())
    clfs.append(DT_classifier)
print(
    "Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
        clfs[-1].tree_.node_count, ccp_alphas[-1]
    )
)
```

```python
clfs = clfs[:-1]
new_ccp_alphas = new_ccp_alphas[:-1]
```

```python
node_counts = [clf.tree_.node_count for clf in clfs]
depth = [clf.tree_.max_depth for clf in clfs]
fig, ax = plt.subplots(2, 1)
fig = plt.figure(figsize =(12, 7))
ax[0].plot(new_ccp_alphas, node_counts, marker="o", drawstyle="steps-post")
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("number of nodes")
ax[0].set_title("Number of nodes vs alpha")
ax[1].plot(new_ccp_alphas, depth, marker="o", drawstyle="steps-post")
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("depth of tree")
ax[1].set_title("Depth vs alpha")
fig.tight_layout()
```

```python
train_scores = [clf.score(X_train, y_train) for clf in clfs]
test_scores = [clf.score(X_test, y_test) for clf in clfs]

fig = plt.figure(figsize =(12, 7))
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(new_ccp_alphas, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(new_ccp_alphas, test_scores, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
#DT_classifier_post_pruning = DecisionTreeClassifier(max_depth= 4, min_samples_leaf= 3, min_samples_split= 5, criterion='entropy')
DT_classifier_post_pruning = DecisionTreeClassifier(ccp_alpha=0.01)
DT_classifier_post_pruning.fit(X_train, y_train.ravel())

y_pred_post_pruning = DT_classifier_post_pruning.predict(X_train)
print(confusion_matrix(y_train,y_pred_post_pruning))
print(classification_report(y_train,y_pred_post_pruning))
print('Decision Tree accuracy: ', accuracy_score(y_train, y_pred_post_pruning).round(2))
```
